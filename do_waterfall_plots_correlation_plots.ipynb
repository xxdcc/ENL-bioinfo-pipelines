# 1. Waterfall plots 
Same as you did before: pair-wise compare T vs WT, first generate a waterfall plot to rank peaks in WT sample by ChIP signal, set the cutoff at where the curve turns, dots below the cutoff use grey color, above the cutoff use purple (to avoid using red again). 

```python
def prepare_waterfall_plot(df, colname='293_F_ENL_treat'):
    df_sort  = df.sort_values(by=[colname])
    df_sort.index = np.arange(df_sort.shape[0])
    return df_sort
```

```python
# find the counts files
files_to_waterfall = glob.glob(sigpath + '*merge.counts.anno.txt')
files_to_waterfall.sort()
print('\n'.join(files_to_waterfall))
print('-----------------')
```

```python
infile = files_to_waterfall[0]
df = pd.read_csv(infile, sep='\t')
col_to_sort = '293_F_ENL_treat'
df_sort = prepare_waterfall_plot(df, colname=col_to_sort)
df_sort.head()
wt_sig = df_sort[col_to_sort].values
clip_val = 80000
fig, ax = plt.subplots(figsize=(5,4))
ax.scatter(np.arange(len(wt_sig)), wt_sig, color='grey', alpha=0.3)
ax.set_xlabel('Rank from small to large')
ax.set_ylabel('WT ChIP signal')
ax.axhline(y=clip_val, linewidth=2, linestyle='--', color='purple', alpha=0.5)

# print the points > clip_val using red color
wt_high = df_sort[wt_sig > clip_val]
ax.scatter(wt_high.index, wt_high[col_to_sort].values, color='purple', alpha=0.6)
ax.set_yticks([0,clip_val,200000,400000,600000,800000,1000000])
name = infile.split('/')[-1].split('.')[0]
ax.set_title(name)
plt.savefig(f'{name}_waterfall.pdf')
```

```python
fig, axes = plt.subplots(4,1, figsize=(5,20), sharex=False)
for i, infile in enumerate(files_to_fc):
    col_to_sort = '293_F_ENL_treat'
    df = pd.read_csv(infile, sep='\t')
    print(infile)
    df['fold_change'] = df.iloc[:,4].div(df.iloc[:,5])
    df_sort = prepare_waterfall_plot(df, colname=col_to_sort)
#     print(df_sort.iloc[:6,:])
    axes[i].scatter(df_sort.index, df_sort.fold_change, color='#DDDDDD', marker='o', alpha=0.5)
    axes[i].set_xlabel('WT ChIP signal rank')
    axes[i].set_ylabel(f'Fold change of Mutant v.s. WT')
    name = infile.split('/')[-1].split('.')[0]
    axes[i].set_title(name)
    # plot the red dots for higher signal in waterfall plot
    wt_high = df_sort[df_sort[col_to_sort] > clip_val]
    axes[i].scatter(wt_high.index, wt_high.fold_change, color='purple', marker='o', alpha=0.5)

plt.savefig('293_fc_wt.pdf')    
```

```python
fig, axes = plt.subplots(4,1, figsize=(5,20), sharex=False)
for i, infile in enumerate(files_to_fc):
    col_to_sort = '293_F_ENL_treat'
    df = pd.read_csv(infile, sep='\t')
    df['fold_change'] = df.iloc[:,4].div(df.iloc[:,5])
    print(infile)
    df_sort = prepare_waterfall_plot(df, colname=col_to_sort)
    axes[i].scatter(df_sort.index, np.log2(df_sort.fold_change+0.01), color='#DDDDDD', marker='o', alpha=0.5)
    axes[i].set_xlabel('WT ChIP signal rank')
    axes[i].set_ylabel(f'Log fold change of Mutant v.s. WT')
    name = infile.split('/')[-1].split('.')[0]
    axes[i].set_title(name)
    # plot the red dots for higher signal in waterfall plot
    wt_high = df_sort[df_sort[col_to_sort] > clip_val]
    axes[i].scatter(wt_high.index, np.log2(wt_high.fold_change+0.01), color='purple', marker='o', alpha=0.5)
    
plt.savefig('293_log2fc_wt.pdf')
```
